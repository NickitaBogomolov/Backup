```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class TextClassifier(nn.Module):
    def __init__(self, input_dim, hidden_dim, num_classes):
        super(TextClassifier, self).__init__()
        self.fc1 = nn.Linear(input_dim, hidden_dim)
        self.fc2 = nn.Linear(hidden_dim, hidden_dim // 2)
        self.fc3 = nn.Linear(hidden_dim // 2, num_classes)
        self.dropout = nn.Dropout(0.5)

    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = self.dropout(x)
        x = F.relu(self.fc2(x))
        x = self.dropout(x)
        x = self.fc3(x)
        return x

# Пример использования:
input_dim = 1024  # Размерность эмбеддингов
hidden_dim = 512  # Размерность скрытого слоя
num_classes = 3   # Количество классов
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
device
```

```python
model = TextClassifier(input_dim, hidden_dim, num_classes)
print(model)
```

```python
from sklearn.metrics import accuracy_score
```

```python
batch_size = 32
input_data = embeddings
labels = torch.from_numpy(np.array(data_4k['class']))
```

```python
torch.from_numpy(np.array(data_4k['class']))
```

```python
# Функция потерь и оптимизатор
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Обучение модели
num_epochs = 10
for epoch in range(num_epochs):
    model.train()  # Переводим модель в режим обучения
    optimizer.zero_grad()

    # Прямой проход
    outputs = model(input_data)
    loss = criterion(outputs, labels)

    # Обратный проход и оптимизация
    loss.backward()
    optimizer.step()

    # Вычисляем accuracy
    model.eval()  # Переводим модель в режим оценки
    with torch.no_grad():
        predictions = torch.argmax(outputs, dim=1)  # Получаем предсказанные классы
        accuracy = accuracy_score(labels.cpu().numpy(), predictions.cpu().numpy())

    # Выводим результаты
    print(f'Epoch {epoch+1}/{num_epochs}, Loss: {loss.item():.4f}, Accuracy: {accuracy:.4f}')
```
